```julia
main_dir = joinpath("output_sorted", "different_tranges")

dirs = readdir(main_dir)

# Initialize dictionaries to hold each column's data
t_data = Dict{Int, Vector}()
It_data = Dict{Int, Vector}()
trange = Dict{Int, Int}()


λ_data = Dict{Int, Vector}()
Iλ_data = Dict{Int, Vector}()
ϕω_data = Dict{Int, Vector}()

# Iterate over each entry in the folder
for dir in dirs
    # Construct the full path of the entry
    full_path = joinpath(main_dir, dir)

    # pulse temporal
    file = joinpath(full_path, "time_domain_output.txt")
    data = readdlm(file, '\t', skipstart=1) 

    Nt = length(data[:,1])
    t_data[Nt] = data[:,1]
    It_data[Nt] = data[:,2]
    trange[Nt] = round(t_data[Nt][end]-t_data[Nt][1])

    #spectrum
    file = joinpath(full_path, "output_spectrum.txt")
    data = readdlm(file, '\t', skipstart=1)

    # Nω = length(data[:,1])
    # println(Nω)

    λ_data[Nt] = data[:,1]   
    Iλ_data[Nt] = data[:,2]
    ϕω_data[Nt] = data[:,3]

    
end
```

```julia
function extract_parameter(file_path::String, parameter_name::String)
    # Initialize a variable to hold the parameter value
    parameter_value = nothing

    # Open and read the file
    open(file_path, "r") do file
        for line in eachline(file)
            # Split the line at the equals sign, and strip whitespaces around each part
            parts = split(strip(line), "=")
            if length(parts) == 2
                name, value_str = strip(parts[1]), strip(parts[2])
                # Check if the left side matches the parameter name
                if name == parameter_name
                    try
                        # Try to parse it as a number (Int or Float64)
                        parameter_value = parse(Float64, value_str)
                    catch e
                        # If parsing fails, it might be a String
                        parameter_value = value_str
                    end
                    break
                end
            end
        end
    end

    return parameter_value
end
```

```julia
dirs = readdir(main_dir)

# Initialize dictionaries to hold each column's data
Npass = Dict{Int, Vector}()
w0 = Dict{Int, Vector}()
wm = Dict{Int, Vector}()


# Iterate over each entry in the folder
for dir in dirs
    # Construct the full path of the entry
    full_path = joinpath(main_dir, dir)

    N_Hankel = Int(extract_parameter(joinpath(full_path, "params.txt"), "N_hankel"))
    println(N_Hankel)

    # beam waist center
    file = joinpath(full_path, "beam_sizes_center.txt")
    data = readdlm(file, '\t', skipstart=1) 

    Npass[N_Hankel] = data[:,1]
    w0[N_Hankel] = data[:,2].*1e6 


    # beam waist mirror position
    file = joinpath(full_path, "beam_sizes_mirror.txt")
    data = readdlm(file, '\t', skipstart=1) 

    Npass[N_Hankel] = data[:,1]
    wm[N_Hankel] = data[:,2].*1e6 
  
end
```

```julia
plt.figure()
plt.title("Beam waist center position")
plt.plot(Npass[32], w0[32], label="32")
plt.plot(Npass[64], w0[64], label="64")
plt.plot(Npass[128], w0[128], label="128")
plt.plot(Npass[256], w0[256], label="256")
plt.plot(Npass[512], w0[512], label="512")
plt.plot(Npass[1024], w0[1024], label="1024")
plt.plot(Npass[2048], w0[2048], label="2048")

plt.xlabel("pass number")
plt.ylabel("w0 [um]")
plt.legend(title="r #samples", loc="upper right")
plt.gcf()
```
